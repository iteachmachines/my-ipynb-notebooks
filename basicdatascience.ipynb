```python
#Scope -There are three  types local,global and built in scope
x=2
def f():
    x=3
    return x
print(x) #global scope
print(f()) #local scope 
```

```python
#Nested function -Function inside function 
def square():
    def add():
        x=2
        y=2
        z=x+y
        return z
    return add()**2
print (square())
```

```python
#Lambda function -Faster way of writing functions
square=lambda x: x**2
print(square(4))
tot=lambda x,y,z:x+y+z
print(tot(1,2,3))
```

```python
#List comprehension
num1=[1,2,3]
num2=[i + 1 for i in num1]
print(num2)
```
